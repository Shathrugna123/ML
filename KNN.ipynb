```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Sample Data (X: features, y: labels)
# Class 0 = Red Circles (🟥), Class 1 = Blue Squares (🟦)
X = np.array([[1, 2], [2, 3], [3, 1], [6, 5], [7, 7], [8, 6]])
y = np.array([0, 0, 0, 1, 1, 1])

# New point to classify
new_point = np.array([[5, 5]])

# Create KNN model with K=3
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

# Predict the label for new point
prediction = model.predict(new_point)
print("Predicted Class:", prediction[0])

# Plotting
plt.figure(figsize=(8, 6))

# Plot existing points
for i, label in enumerate(y):
    if label == 0:
        plt.scatter(X[i, 0], X[i, 1], c='red', marker='o', label='Class 0' if i == 0 else "")
    else:
        plt.scatter(X[i, 0], X[i, 1], c='blue', marker='s', label='Class 1' if i == 3 else "")

# Plot new point
plt.scatter(new_point[0][0], new_point[0][1], c='green', marker='X', s=200, label='New Point (?)')

# Set plot details
plt.title('KNN (K=3) Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
